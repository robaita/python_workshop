<h1>Introduction to Pandas</h1>
<h2>What is Pandas? </h2>
Pandas is an open-source data analysis and manipulation library in Python, widely used in data science, machine learning, and data engineering. It provides data structures and functions needed to efficiently work with structured data, especially tabular data (like data from databases or spreadsheets).

<h2>Importance of Pandas</h2>
<ul>
<li><strong>Efficient Data Handling:</strong> Makes it easy to clean, filter, manipulate, and analyze data.</li>
<li><strong>Powerful Data Analysis:</strong> Offers robust tools to perform operations on large datasets, making complex transformations easy.</li>
<li><strong>High Compatibility: </strong>Works well with other data analysis libraries and integrates with databases, files, and visualization libraries like Matplotlib.</li>
</ul>

<h2>How Pandas Works </h2>
Pandas primarily works with two data structures:
<ul>
<li><strong>Series: </strong>A one-dimensional labeled array, ideal for simple lists or arrays.</li>
<li><strong>DataFrame:</strong> A two-dimensional labeled data structure with columns of potentially different types, akin to a table in a database or an Excel sheet.</li>
</ul>

<h1>Pandas Dataframe </h1>
A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns). Think of it as a powerful version of an Excel spreadsheet or SQL table that allows for easy data manipulation and analysis. A DataFrame is one of the core data structures in Pandas, alongside the Series, and it is widely used for data analysis in Python.

<h2>Key Properties of a DataFrame</h2>
<ul>
<li><strong>Two-Dimensional Structure:</strong> DataFrames are structured as a table with rows and columns, where each column can hold data of a different type (integer, float, string, etc.).</li>

<li><strong>Labeled Axes: </strong>Both rows and columns in a DataFrame are labeled, making it easy to access data by row/column names or indexes.</li>

<li><strong>Mutable Size:</strong> DataFrames can grow or shrink in size. You can add or remove rows and columns as needed.</li>

<li><strong>Heterogeneous Data:</strong> Unlike arrays, a DataFrame can store data of different types in different columns, allowing flexibility in storing and manipulating data from real-world applications.</li>

<li><strong>Data Alignment:</strong> When performing operations on DataFrames, Pandas aligns data based on row and column labels, making it easier to work with data from different sources or perform complex calculations.</li>

<li><strong>Integration with Other Libraries:</strong> Pandas works well with other Python libraries like NumPy, Matplotlib, and Scikit-Learn, providing a strong ecosystem for data analysis and machine learning.</li>
</ul>

<h2>Basic Structure of a DataFrame</h2>

<ul>
<li><strong>Index (rows):</strong> Uniquely identifies each row. Default is a numerical index, but custom indexes can also be used.</li>
<li><strong>Columns: </strong>Each column has a name (header) and a data type (e.g., integer, float, string).</li>
</ul>

<div id="BBox" class="alert alert-info" style="font-family:courier;color:black;justify-content:left;">
<h1>df.pivot_table()</h1>
<strong>What it does:</strong> Creates a pivot table, summarizing data by groups.<br>
<strong>How it works: </strong>  Aggregates values across specified indexes and columns.<br>
<strong>When to use:</strong> For advanced data summaries, especially in analysis.<br><br>
</div>

```python
data = {
    'Salesperson': ['Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob'],
    'Region': ['North', 'North', 'South', 'South', 'North', 'South'],
    'Product': ['Widgets', 'Widgets', 'Gadgets', 'Widgets', 'Gadgets', 'Gadgets'],
    'Sales': [200, 150, 300, 250, 400, 350]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

pivot = df.pivot_table(
    values='Sales',            # Column to aggregate
    index='Salesperson',       # Rows
    columns=['Region', 'Product'], # Columns to pivot
    aggfunc='sum',             # Aggregation function
    fill_value=0               # Replace missing values with 0
)

print("\nPivot Table:")
print(pivot)

```

```python
new_df = pd.read_csv("data\world_population.csv")
new_df.rename(columns={'Country/Territory': 'Country'}, inplace=True)
new_df = new_df[['Country','Continent', '2022 Population']]
pivot = new_df.pivot_table(
    values='2022 Population',            # Column to aggregate
    columns=['Continent'], # Columns to pivot
    aggfunc='sum',             # Aggregation function
    fill_value=0               # Replace missing values with 0
)

print("\nPivot Table:")
pivot.head(10)
```

```python
import matplotlib.pyplot as plt

data = pd.read_csv("data/world_population.csv")
data.plot.line(x='CCA3', y='2022 Population')
plt.title("Country VS 2022 Population")
plt.show()
```

```python
data.groupby("Continent")["2022 Population"].sum().plot.bar()
plt.title("Population by Continent")
plt.show()
```

```python
data.groupby("CCA3")["2022 Population"].sum().nlargest(10).plot.barh()
plt.title("Top 10 Most Populous Countries")
plt.show()
```

```python
data["Area (km2)"].plot.hist(bins=30)
plt.title("Area Distribution")
plt.xlabel("Area")
plt.show()
```

```python
data.groupby("Continent")["2022 Population"].sum().plot.pie(autopct='%1.1f%%')
plt.title("Population by Continent")
plt.ylabel("")
plt.show()
```
