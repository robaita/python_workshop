<h1>Line Plot</h1>
<strong>What it does:</strong>
A line plot displays data points connected by straight line segments. Itâ€™s commonly used to visualize the relationship between two continuous variables, especially trends over time. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, the plot() function connects the points, forming a line that visualizes the relationship between x and y.
</span>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 10, 100)
y = 4 + 1 * np.sin(2 * x)
y2 = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y2 + 2.5, 'g.', markeredgewidth=2)
ax.plot(x, y, 'c-',linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 10), yticks=np.arange(1, 10))

plt.show()
```
