<h1>1. Line Plot</h1>
<strong>What it does:</strong>
A line plot displays data points connected by straight line segments. Itâ€™s commonly used to visualize the relationship between two continuous variables, especially trends over time. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, the plot() function connects the points, forming a line that visualizes the relationship between x and y.
</span>
</div>

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Creating the line plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>2. Scatter Plot</h1>
<strong>What it does:</strong>
A scatter plot is used to display values for typically two variables by plotting data points. Itâ€™s useful to observe relationships or clusters in data. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, scatter() creates a plot with scattered points, helping to visualize the spread and correlation between x and y.
</span>
</div>

```python
# Data
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

# Creating the scatter plot
plt.scatter(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>3. Bar Plot</h1>
<strong>What it does:</strong>
A bar plot represents categorical data with rectangular bars. The height of each bar is proportional to the values it represents. It's ideal for comparing different groups or categories. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
This example shows a vertical bar plot where each bar represents a category and its corresponding value.
</span>
</div>

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]

# Creating the bar plot
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>4. Histogram</h1>
<strong>What it does:</strong>
A histogram is used to represent the distribution of a dataset by dividing the data into bins. Itâ€™s useful for understanding the frequency distribution of continuous data. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
Here, hist() creates a histogram with 5 bins, illustrating the frequency distribution of the data values.
</span>
</div>

```python
# Data
data = [22, 55, 62, 45, 21, 22, 34, 42, 55, 58, 60, 72, 85]

# Creating the histogram
plt.hist(data, bins=5)

# Adding labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>5. Pie Chart</h1>
<strong>What it does:</strong>
A pie chart displays data as a circular graph, with each slice representing a proportion of the whole. It's useful for showing percentage or proportional data. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, the pie() function creates a pie chart where each slice corresponds to a category and its percentage of the total.
</span>
</div>

```python
# Data
labels = ['A', 'B', 'C', 'D']
sizes = [30, 40, 20, 10]

# Creating the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Adding title
plt.title('Pie Chart Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>6. Box Plot</h1>
<strong>What it does:</strong>
A box plot (or box-and-whisker plot) is used to display the distribution of data and identify outliers. It shows the median, quartiles, and possible outliers of the dataset. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
This example uses boxplot() to visualize the spread of the data values, with the box representing the interquartile range (IQR).
</span>
</div>

```python
# Data
data = [7, 12, 14, 15, 16, 18, 21, 22, 23, 24, 29]

# Creating the box plot
plt.boxplot(data)

# Adding title
plt.title('Box Plot Example')

# Display the plot
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>7. Heatmap (using imshow)</h1>
<strong>What it does:</strong>
A heatmap represents data in matrix form with different colors indicating the intensity of the values. It's used for visualizing complex datasets, especially in correlation matrices or for visualizing clustering. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, imshow() is used to create a heatmap from a 5x5 random matrix, with color representing the intensity of each value.
</span>
</div>

```python
import numpy as np

# Data
matrix_data = np.random.rand(5, 5)

# Creating the heatmap
plt.imshow(matrix_data, cmap='hot', interpolation='nearest')

# Adding title
plt.title('Heatmap Example')

# Display the plot
plt.colorbar()
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>8. Subplots</h1>
<strong>What it does:</strong>
Subplots allow multiple plots to be displayed in a grid. Itâ€™s useful for comparing different plots side by side. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, subplots() creates a 2x2 grid of plots, each demonstrating a different type of visualization.
</span>
</div>

```python
# Creating subplots
fig, axs = plt.subplots(2, 2)

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Top-left subplot (Line plot)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')

# Top-right subplot (Bar plot)
axs[0, 1].bar(['A', 'B', 'C', 'D'], [3, 7, 1, 5])
axs[0, 1].set_title('Bar Plot')

# Bottom-left subplot (Scatter plot)
axs[1, 0].scatter([1, 2, 3, 4], [10, 15, 13, 20])
axs[1, 0].set_title('Scatter Plot')

# Bottom-right subplot (Histogram)
axs[1, 1].hist([1, 2, 2, 3, 3, 3, 4, 4, 5])
axs[1, 1].set_title('Histogram')

# Adjust layout
plt.tight_layout()

# Display the subplots
plt.show()
```

<h1>Let's make some complex graphs</h1>
<strong>Multi Line plot</strong>
<span style="font-family: Courier New, monospace; color: black;">
The plt.text() function is used to place text annotations on the graph. <br>
For each share, the value is placed at the respective data points (time[i], share[i]).<br>
Colors are chosen for better visibility (blue for Share 1, green for Share 2, and red for Share 3).<br><br>

In the plt.text() function, the parameters ha and va control the horizontal and vertical alignment of the text relative to the point where it is plotted.<br>

ha (horizontal alignment): It controls the horizontal positioning of the text relative to the specified x coordinate. Possible values are:<br>
'center': Aligns the center of the text with the x position.<br>
'left': Aligns the left edge of the text with the x position.<br>
'right': Aligns the right edge of the text with the x position.<br><br>

va (vertical alignment): It controls the vertical positioning of the text relative to the specified y coordinate. Possible values are:<br>
'center': Aligns the middle of the text with the y position.<br>
'top': Aligns the top of the text with the y position.<br>
'bottom': Aligns the bottom of the text with the y position.<br>
'baseline': Aligns the text's baseline with the y position.
    
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example time data (e.g., days 1 to 10)
time = np.arange(1, 11)

# Example share prices for three different shares
share1 = [100, 102, 105, 103, 108, 110, 107, 111, 115, 117]
share2 = [98, 99, 101, 103, 106, 107, 109, 108, 110, 112]
share3 = [105, 107, 106, 110, 112, 111, 113, 116, 118, 120]

# Annotate each point with the share price
for i in range(len(time)):
    plt.text(time[i], share1[i], f'{share1[i]}', ha='center', va='bottom', fontsize=9, color='blue')
    plt.text(time[i], share2[i], f'{share2[i]}', ha='center', va='bottom', fontsize=9, color='green')
    plt.text(time[i], share3[i], f'{share3[i]}', ha='center', va='bottom', fontsize=9, color='red')

# Create the line plot
plt.plot(time, share1, label='Share 1', marker='o')
plt.plot(time, share2, label='Share 2', marker='x')
plt.plot(time, share3, label='Share 3', marker='s')

# Adding labels and title
plt.xlabel('Time (Days)')
plt.ylabel('Share Price ($)')
plt.title('Share Price Over Time')

# Show legend to differentiate between shares
plt.legend()

# Display the plot
plt.show()
```
