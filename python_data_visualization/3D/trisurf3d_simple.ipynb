<h1>TriSurf Plot</h1>
The plot_trisurf function in Matplotlib is used to create a three-dimensional triangular surface plot. This function is particularly useful when you have scattered data points that do not form a regular grid, as it can interpolate a surface based on the points provided. It's commonly used in fields such as geology, meteorology, and engineering to visualize complex surfaces.<br><br>
<strong>Key Features of plot_trisurf:</strong>
<ul>
  <li>Triangulation of Irregular Data: Unlike plot_surface, which requires a grid of points, plot_trisurf can handle scattered data points by triangulating them.</li>
  <li>Flexible Input: You can provide a set of x, y, and z coordinates that represent the vertices of the surface.</li>
  <li>Customization: It allows for customization of the color, shading, and edge properties of the triangular facets.</li>
</ul>

<strong>Use Case:</strong>
<ul>
  <li>Geospatial Data: Useful for visualizing terrain data where measurements are taken at irregular intervals.</li>
  <li>Computational Fluid Dynamics: Helps in representing flow fields where data points are not evenly spaced.</li>
  <li>Finite Element Analysis: Commonly used in engineering simulations to visualize the results of mesh-based computations.</li>
</ul> 

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

n_radii = 8
n_angles = 36

# Make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
