
# plot(xs, ys, zs)

See `~mpl_toolkits.mplot3d.axes3d.Axes3D.plot`.


```python
import matplotlib.pyplot as plt
import numpy as np

# Make data
n = 100
xs = np.linspace(0, 1, n)
ys = np.sin(xs * 6 * np.pi)
zs = np.cos(xs * 6 * np.pi)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
