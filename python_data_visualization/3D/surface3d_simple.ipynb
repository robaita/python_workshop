<h1>Surface Plot</h1>
The plot_surface function in Matplotlib is used to create a three-dimensional surface plot. This type of plot is particularly useful for visualizing data that varies over two dimensions, allowing you to see the shape and features of the surface. It is commonly used in scientific and engineering applications where the relationship between three variables needs to be illustrated.<br><br>
<strong>Key Features of plot_surface:</strong>
<ul>
  <li>3D Visualization: Represents data in three dimensions, which can reveal patterns that are not visible in 2D plots.</li>
  <li>Continuous Data Representation: Ideal for visualizing continuous functions or datasets where values change across two dimensions.</li>
  <li>Customizable Appearance: You can customize the colors, shading, and transparency of the surface to enhance the visualization.</li>
</ul>
    
<strong>Explanation:</strong>
<ul>
  <li>np.meshgrid, which generates a grid of points.</li>
  <li>R is calculated using a mathematical function (sin(sqrt(x^2 + y^2))) that defines the surface.</li>
  <li>projection='3d', initializes a 3D axis for plotting.</li>
  <li>Color Map: The cmap parameter specifies the color map to use for the surface, enhancing visual appeal and clarity.</li>
</ul>    

<strong>Use Case:</strong>
<ul>
  <li>Scientific Visualization: Useful in fields such as physics, engineering, and meteorology to visualize surfaces like potential energy fields or temperature distributions.</li>
  <li>Data Analysis: Helpful in data analysis and machine learning for visualizing loss functions or model performance over different parameters.</li>
  <li>Mathematical Functions: Ideal for visualizing complex mathematical functions in three dimensions.</li>
</ul> 

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
