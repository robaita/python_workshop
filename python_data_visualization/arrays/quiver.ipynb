<h1>Quiver Plot</h1>
A quiver plot is a type of plot used to represent vector fields by drawing arrows (quivers) at points in a grid. Each arrow represents a vector, with its direction indicating the direction of the vector and its length representing the magnitude.

Quiver plots are commonly used in physics and engineering to display velocity fields, force fields, and other vector quantities that vary over a two-dimensional space.<br><br>
    
<strong>Key Components of a Quiver Plot:</strong> <br>
<ul>
  <li>Arrow direction: Represents the direction of the vector at each point..</li>
  <li>Arrow length: Represents the magnitude of the vector at that point.</li>
  <li>Grid of points: The arrows are placed on a grid, which could be evenly spaced or based on specific data.</li>
</ul>
    
<strong>Explanation:</strong>
<ul>
  <li>X, Y: These are the grid coordinates generated using np.meshgrid().</li>
  <li>U, V: These are the vector components at each point (X, Y). In this example, U and V are simple mathematical functions of X and Y to illustrate the concept.</li>
  <li>plt.quiver(): This function generates the quiver plot, where arrows are placed at grid points (X, Y) with the direction and magnitude determined by (U, V).</li>
</ul>    

<strong>Use Case:</strong>
   <ul>
      <li>Velocity Fields: Quiver plots are often used to display wind or fluid flow patterns.</li>
      <li>Electromagnetic Fields: Used to visualize electric or magnetic field vectors in physics.</li>
      <li>Force Fields: Can represent force vectors acting on objects in a 2D plane.</li>
      <li>Gradient Visualization: In optimization and machine learning, quiver plots can be used to visualize gradients or vector fields.</li>
   </ul>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(-4, 4, 6)
y = np.linspace(-4, 4, 6)
X, Y = np.meshgrid(x, y)
U = X + Y
V = Y - X

# plot
fig, ax = plt.subplots()

ax.quiver(X, Y, U, V, color="C0", angles='xy',
          scale_units='xy', scale=5, width=.015)

ax.set(xlim=(-5, 5), ylim=(-5, 5))

plt.show()
```
