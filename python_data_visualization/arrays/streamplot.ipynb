<h1>Stream Plot</h1>
A stream plot is a type of plot used to visualize vector fields in a smooth, continuous manner. It draws lines (streamlines) that represent the flow of a vector field, showing the direction of the flow at each point in the field. Stream plots are particularly useful for visualizing fluid dynamics, air flow, and other types of vector fields where it's important to understand how the vectors are moving or flowing through space.<br><br>
    
<strong>Key Components of a Stream Plot:</strong> <br>
<ul>
  <li>Streamlines: Curved lines that follow the direction of the vector field. They give a sense of how the field "flows" over the plotted area.</li>
  <li>Vector direction: The direction of the streamlines indicates the direction of the vector at any point.</li>
  <li>Vector density: Streamlines may be denser in areas where the vector field is stronger, showing areas of greater magnitude.</li>
  <li></li>
</ul>
    
<strong>Difference Between Stream Plot and Quiver Plot:</strong>
<ul>
  <li>A quiver plot shows discrete arrows at specific grid points representing direction and magnitude.</li>
  <li>A stream plot draws continuous lines that flow smoothly through the vector field, showing the overall structure and direction of the field in a more connected way.</li>
</ul>    

<strong>Use Case:</strong>
   <ul>
      <li>Fluid Dynamics: Visualize the flow of liquids or gases in different regions.</li>
      <li>Weather Data: Show wind patterns or ocean currents.</li>
      <li>Magnetic Fields: Illustrate the continuous flow of magnetic fields around magnets or electrical currents.</li>
      <li>Electromagnetic Fields: Visualize electric fields or potential fields in physics.</li>
   </ul>

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```
