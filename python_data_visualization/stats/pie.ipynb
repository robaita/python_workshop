<h1>Pie Chart</h1>
<strong>What it does:</strong>
A pie chart displays data as a circular graph, with each slice representing a proportion of the whole. It's useful for showing percentage or proportional data. <br><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">
In this example, the pie() function creates a pie chart where each slice corresponds to a category and its percentage of the total.
</span>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
