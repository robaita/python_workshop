<h1>1. Seaborn</h1>
Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.<br><br>
<strong>1.1 Scatter Plot</strong><br>
<span style="font-family: Courier New, monospace; color: navy; font-weight: bold;">A scatter plot displays values for typically two variables for a set of data, showing how much one variable is affected by another. </span>
</div>

```python
# !pip install seaborn
# !pip install --upgrade scipy
# !pip install numpy --upgrade
# !pip install --upgrade seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Ensure 'time' is treated as a categorical variable
tips['time'] = tips['time'].astype('category')

print(tips.columns)
# Create a scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>1.2 Line Plot</h1>
What it does: A line plot is used to show trends over time or continuous data.<br><br>
</div>

```python
# Create a line plot
sns.lineplot(data=tips, x="size", y="total_bill", estimator='mean')
plt.title("Line Plot of Average Total Bill by Size")
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>1.3 Box Plot</h1>
What it does: A box plot summarizes the distribution of a dataset based on a five-number summary ("minimum", first quartile (Q1), median, third quartile (Q3), and "maximum").
</div>

```python
# Create a box plot
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Box Plot of Total Bill by Day and Gender")
plt.show()
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>1.4 Heat Plot</h1>
1.4 Heatmap
What it does: A heatmap visualizes data through variations in color, allowing for easy identification of patterns.
</div>

```python
# Create a pivot table for the heatmap
pivot_table = tips.pivot_table(values='total_bill', index='day', columns='time', aggfunc='mean')

# Create a heatmap
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title("Heatmap of Average Total Bill by Day and Time")
plt.show()
```

<h1>2. GGplot</h1>
ggplot is an R package for creating graphics based on the Grammar of Graphics. In Python, you can use the plotnine library to create similar plots. <br>

<h2>2.1 Scatter Plot </h2>
What it does: Similar to Seaborn, it displays the relationship between two continuous variables.
</div>

```python
# !pip install plotnine
```

```python
from plotnine import ggplot, aes, geom_point, ggtitle

# Create a scatter plot
scatter_plot = (ggplot(tips, aes(x='total_bill', y='tip', color='time')) +
                 geom_point() +
                 ggtitle('Scatter Plot of Total Bill vs Tip'))
print(scatter_plot)
```

<div id="BBox" class="alert alert-success" style="font-family:courier;color:black;justify-content:left;">
<h1>2.2 Box Plot</h1>
What it does: Similar to Seaborn, it summarizes data distribution.
</div>

```python
from plotnine import geom_boxplot, ggplot, ggtitle

# Create a box plot
box_plot = (ggplot(tips, aes(x='day', y='total_bill', fill='sex')) +
            geom_boxplot() +
            ggtitle('Box Plot of Total Bill by Day and Gender'))
print(box_plot)
```
